# **How do I rename columns in pandas DataFrame?**

```
ufo.head()
```

```
# List the columns name
ufo.columns
```

```
# Method 1 to change column name

ufo.rename(columns = {'Colors Reported' : 'Colors_Reported', 'Shape Reported' : 'Shape_Reported'}, inplace = True)
```

```
ufo.columns
```

```
# Method 2 to change column name

col = ['city', 'colors reported', 'shape reported', 'state', 'time', 'location']
ufo.columns = col
```

```
ufo.head()
```

```
# Method 3 to change when reading the file

col = ['city', 'colors reported', 'shape reported', 'state', 'time']

ufo = pd.read_csv('http://bit.ly/uforeports', names = col, header = 0)

ufo.head()
```

```
ufo.columns = ufo.columns.str.replace(' ', '_')
```

```
ufo.head()
```

# **How do I remove columns from pandas DataFrame**

```
ufo.head()
```

```
# Dropping single column
ufo.drop('Colors Reported', axis = 1, inplace = True)
# axis = 1 means column
ufo.head()
```

```
# Dropping multiple columns
ufo.drop(['Colors Reported', 'Time'], axis = 1, inplace = True)
ufo.head()
```

```
# Removing rows
ufo.drop([0,1] , axis = 0, inplace = True)
# axis = 0 is default
ufo.head()
```

```
# Reload the original data frame
ufo = pd.read_csv('http://bit.ly/uforeports')
```

# **How do I sort pandas DataFrame or Series**

```
movies.head()
```

```
# Sorting in ascending order
movies['title'].sort_values()
```

```
movies.head()
```

```
# Sorting in descending order
movies['title'].sort_values(ascending = False)
```

```
#Sorting the DataFrame and not keeping the change
movies.sort_values('title', ascending = False)
```

```
# Sorting the DataFrame and keeping the change
movies.sort_values('duration', inplace = True)
movies.head()
```

```
movies.sort_values(['genre', 'duration'])
```
