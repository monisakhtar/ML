## The Long Method

```
# The long method
booleans = []
for length in movies['duration']:
  if length >= 200:
    booleans.append(True)
  else:
    booleans.append(False)
```

```
# Looking at 5 values in booleans
booleans[0:5]
```

```
#Printing my findings
movies[booleans]
```

## The short method

```
# Easy method way 1
booleans = []
# Pandas knows that condition is being checked on all of the named series
booleans = movies['duration'] >= 200
booleans[0:5]
```

```
movies[movies['duration'] >= 200]
```

```
# Pulling out a particular column with a condition
movies[movies['duration'] >= 200]['genre']
# This works only because movies[movies['duration'] >= 200] is a DataFrame
```

## How to chain string method together
## For examples I want to remove the '[' & **']'** from the choice_description. how do I do that?



1.   First I remove '[' this would give a series as an output
2.   On that output we run another str method to remove ']'



```
# Removing the first '['
orders['choice_description'].str.replace('[', '', regex = True)
```

```
# Attaching another str method on the result of first series
orders['choice_description'].str.replace('[', '', regex = True).str.replace(']', '', regex = True)
```

```
# Using regular expression method
orders['choice_description'].str.replace('[\[\]]', '', regex = True)
```

# **How do I handle missing value in pandas?**

```
ufo.head()
```

```
# Creates a DataFrame with true and false to show something is null or not
ufo.isnull()
```

```
# Gives a sum of null values in each column
ufo.isnull().sum()
```

## Missing value can be handled in many ways. the most common ways include :


*   Drop rows or columns that only have missing values. 
*   Drop rows or columns based on a threshold value.
*   Handle missing value by assigning or finding out the value to be assigned.






```
ufo.shape
```

```
# Drop any row that has a missing value in any of its column
ufo.dropna(how = 'any').shape
```

```
# We need to specify inplace = True to apply changes
ufo.shape
```

```
# drop rows if a row contains only null value
ufo.dropna(how = 'all').shape
```

```
# Drop a row only based on a subset of column
ufo.dropna(subset = ['City', 'Shape Reported'], how = 'all').shape
```

```
# value_counts has not counted the missing values
ufo['Shape Reported'].value_counts()
```

```
# Counting the missing values as well
ufo['Shape Reported'].value_counts(dropna = False)
```

```
# Filling the missing value with "UNKNOWN"
# We can use inplace = True to apply the change
ufo['Shape Reported'].fillna(value = 'UNKNOWN').value_counts(dropna = False)
```

# **How do I create dummy variable in pandas?**

```
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```
train['Is_Male'] = train['Sex'].map({'female': 0, 'male': 1})
```

```
train.head()
```

```
pd.get_dummies(train['Sex'])
```

```
# Represent it with one column
pd.get_dummies(train['Sex']).iloc[:, 1:]
```

```
# Add a prefix to the column in dummy variables
pd.get_dummies(train['Sex'], prefix = 'Is').iloc[:, 1:]
```

```
train['Embarked'].value_counts()
```

```
pd.get_dummies(train['Embarked'], prefix = 'Embarked')
```

```
embarked_dummies = pd.get_dummies(train['Embarked'], prefix = 'Embarked').iloc[:, 1:]
```

```
train = pd.concat([train, embarked_dummies], axis = 1)
```

```
train.head()
```

```
# Trick to create dummy variables for multiple column
pd.get_dummies(train, columns = ['Sex', 'Embarked'])
```

```
# How to exclude first dummy column
pd.get_dummies(train, columns = ['Sex', 'Embarked'], drop_first=True)
```
