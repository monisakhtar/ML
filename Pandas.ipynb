# **How do I filter rows of a pandas DataFrame by column value**

```
movies.head()
```

```
movies.shape
```

## The Long Method

```
# The long method
booleans = []
for length in movies['duration']:
  if length >= 200:
    booleans.append(True)
  else:
    booleans.append(False)
```

```
# Looking at 5 values in booleans
booleans[0:5]
```

```
#Printing my findings
movies[booleans]
```

## The short method

```
# Easy method way 1
booleans = []
# Pandas knows that condition is being checked on all of the named series
booleans = movies['duration'] >= 200
booleans[0:5]
```

```
movies[movies['duration'] >= 200]
```

```
# Pulling out a particular column with a condition
movies[movies['duration'] >= 200]['genre']
# This works only because movies[movies['duration'] >= 200] is a DataFrame
```

# **How do i use string methods in pandas**

```
orders.head()
```

```
# To use string method in pandas you always have to use .str method along with the method we want to use
orders['item_name'].str.upper()
```

```
orders[orders['item_name'].str.contains('Chicken')]
```

## How to chain string method together
## For examples I want to remove the '[' & **']'** from the choice_description. how do I do that?



1.   First I remove '[' this would give a series as an output
2.   On that output we run another str method to remove ']'



```
# Removing the first '['
orders['choice_description'].str.replace('[', '', regex = True)
```

```
# Attaching another str method on the result of first series
orders['choice_description'].str.replace('[', '', regex = True).str.replace(']', '', regex = True)
```

```
# Using regular expression method
orders['choice_description'].str.replace('[\[\]]', '', regex = True)
```

# **When should i use 'groupby' in pandas?**

```
drinks.head()
```

```
#Average beerserving across the dataset or all countries
drinks['beer_servings'].mean()
```

```
# How to find out across each continent
drinks.groupby('continent').mean()
```

```
# How to find out across each continent by a specific column value
drinks.groupby('continent')['beer_servings'].mean()
```

```
# the agg() function lets us groups different function together
drinks.groupby('continent')['beer_servings'].agg(['count', 'min', 'max', 'mean'])
```

```
%matplotlib inline
```

```
drinks.groupby('continent').mean().plot(kind = 'bar')
```
